```python
def get_page_data(page_soup: BeautifulSoup) -> dict:
    data = {}
    script_tags = page_soup.find_all('script')
    
    for tag in script_tags:
        try:
            tag_contents = tag.string.extract()
        except AttributeError:
            continue

        if re.match(r'\s*window\.preLoaded\s=', tag_contents):
            start = tag_contents.index('=') + 2
            end = tag_contents.index('};') + 1
            # print(tag_contents[start:end])
            data = json.loads( tag_contents[start:end] )
            break
        
    return data
```

```python
sneakers_urls = pd.read_csv('../data.csv', index_col=0)[['item_sku', 'item_url']]
```

```python
df = pd.DataFrame()

for i in range(len(sneakers_urls.index)):
    url = sneakers_urls.loc[i, 'item_url']
    print(Fore.BLUE + f'{i + 1}. Getting data for <{url}>...', end=' ')

    page_data = None
    try_num = 1
    while not page_data:
        try:
            page_data = get_page_data(get_page_soup(url))
        except HTTPError:
            print(Fore.RED + f'\n403, waiting {5 * try_num} minutes...', end=' ')
            sleep(5 * 60 * try_num)
            try_num += 1

    df.append(scrape_page_data(page_data), ignore_index=True)
    sleep(5)

    print(Fore.GREEN + f'DONE ({ round(100 * (i + 1) / len(sneakers_urls.index), 2) }%)')

print(Fore.BLUE + 'Saving as "data pages.csv"')
df.to_csv('../data pages.csv')
```
